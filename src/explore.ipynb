```python
import pandas as pd

total_movies = pd.read_csv("https://raw.githubusercontent.com/4GeeksAcademy/k-nearest-neighbors-project-tutorial/main/tmdb_5000_movies.csv")
total_movies.head(5)
```

```python
total_movies['title'].value_counts()
```

```python
total_credits = pd.read_csv("https://raw.githubusercontent.com/4GeeksAcademy/k-nearest-neighbors-project-tutorial/main/tmdb_5000_credits.csv")
total_credits.head(5)
```

```python
total_data = pd.merge(total_movies, total_credits, left_on=['id', 'title'], right_on=["movie_id", "title"])
total_data.head()
```

```python
total_data.columns
```

```python
columns_to_keep = ['movie_id', 'title', 'overview', 'genres', 'keywords', 'cast', 'crew']
total_data = total_data[columns_to_keep]
total_data.head()
```

```python
# Check the data type of `genres`, `keywords`, `cast`, `crew`

columns_to_check = ["genres", "keywords", "cast", "crew"]

for column in columns_to_check:
    print(f"Data type of the first row for column `{column}`: {type(total_data.loc[0, column])}")
```

```python
import ast

# Convert string representation of lists to actual lists
total_data['genres'] = total_data['genres'].apply(lambda x: ast.literal_eval(x) if pd.notnull(x) else x)
total_data['keywords'] = total_data['keywords'].apply(lambda x: ast.literal_eval(x) if pd.notnull(x) else x)
total_data['cast'] = total_data['cast'].apply(lambda x: ast.literal_eval(x) if pd.notnull(x) else x)
total_data['crew'] = total_data['crew'].apply(lambda x: ast.literal_eval(x) if pd.notnull(x) else x)

# Verify that the data type was changed from a str to list
for column in columns_to_check:
    print(f"Data type of the first row for column `{column}`: {type(total_data.loc[0, column])}")
```

```python
print('genres')
total_data['genres'].head()
```

```python
# Complete the `extract_name()` function to extract the 'name' attribute from the JSON

def extract_name(json_list):
    # Initialize an empty list to hold the names
    names = []
    
    # Loop through each item in the json_list
    for item in json_list:
        # Check if the 'name' key exists in the current item
        if 'name' in item:
            # If it exists, append the value of the 'name' key to the names list
            names.append(item['name'])
    
    # Return the list of names after going through all items
    return names
    

# Apply the `extract_name` function to 'genres' and 'keywords' columns
    
total_data['genres'] = total_data['genres'].apply(extract_name)
total_data['keywords'] = total_data['keywords'].apply(extract_name)

def select_first_three_names(json_list):
    # Initialize an empty list to hold the names
    names = []
    
    # Loop through each item in the json_list
    for item in json_list:
        names.append(item['name'])
        # Check if we have already collected three names to stop processing further
        if len(names) == 3:
            break         
    # Return the list of first three names
    return names

# Apply the `select_first_three_names` function to 'cast' column
    
total_data['cast'] = total_data['cast'].apply(select_first_three_names)

total_data.head()
```

```python
first_row_crew = total_data.loc[0, 'crew']

for dict_obj in first_row_crew:
    print(dict_obj)
    print()
```

```python
def select_director_name(json_list):
    name = ''

    # Go through each dictionary in list to find the director's name
    for dict_obj in json_list:
        if dict_obj["job"] == "Director":
            name = dict_obj["name"]

    return name

total_data['crew'] = total_data['crew'].apply(select_director_name)

total_data.head()
```

```python
total_data['genres'] = total_data['genres'].apply(lambda x: ' '.join(x))
total_data['keywords'] = total_data['keywords'].apply(lambda x: ' '.join(x))
total_data['cast'] = total_data['cast'].apply(lambda x: ' '.join(x))
total_data
```

```python
total_data[total_data.isnull().any(axis=1)]
```

```python
total_data['overview'] = total_data['overview'].fillna('')
```

```python
total_data['tags'] = total_data['overview'] + " " + total_data['genres'] + " " + total_data['keywords'] + " " + total_data['cast']

print(total_data.iloc[0].tags)

total_data.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.neighbors import NearestNeighbors

vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(total_data["tags"])

model = NearestNeighbors(n_neighbors = 6, algorithm = "brute", metric = "cosine")
model.fit(tfidf_matrix)

def get_movie_recommendations(movie_title):
    movie_index = total_data[total_data["title"] == movie_title].index[0]
    distances, indices = model.kneighbors(tfidf_matrix[movie_index])
    similar_movies = [(total_data["title"][i], distances[0][j]) for j, i in enumerate(indices[0])]
    return similar_movies[1:]

input_movie = "How to Train Your Dragon"
recommendations = get_movie_recommendations(input_movie)
print("Film recommendations '{}'".format(input_movie))
for movie, distance in recommendations:
    print("- Film: {}".format(movie))
```
